```python
x=cow[['work_year', 'experience_level', 'employment_type', 'job_title',
       'salary', 'salary_currency', 'employee_residence',
       'remote_ratio', 'company_location', 'company_size']]
```

```python
np.random.seed(0)
X = np.random.rand(100, 3)
y = 2.5 + 1.2 * X[:, 0] + 0.8 * X[:, 1] - 1.5 * X[:, 2] + np.random.normal(0, 0.2, 100)
```

```python
X = sm.add_constant(X)
```

```python
models = {
    'model_1': sm.OLS(y, X[:, :2]).fit(),
    'model_2': sm.OLS(y, X[:, :3]).fit(),
    'model_3': sm.OLS(y, X).fit()
}
```

```python
results = []
for name, model in models.items():
    aic = model.aic
    bic = model.bic
    results.append((name, aic, bic))
```

```python
results_df = pd.DataFrame(results, columns=['Model', 'AIC', 'BIC'])
print(results_df)
```
